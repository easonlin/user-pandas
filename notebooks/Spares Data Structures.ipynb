# Sparse data structures http://pandas.pydata.org/pandas-docs/stable/sparse.html

```python
from pandas import *
import numpy as np
ts = Series(randn(10))
ts[2: -2] = np.nan
sts = ts.to_sparse()
sts
```

```python
ts.fillna(0).to_sparse(fill_value=0)
```
