## Covariance

```python
# Excluding None/nan
s1 = Series(randn(1000))
s2 = Series(randn(1000))
s1.cov(s2)
```

```python
frame = DataFrame(randn(1000, 5), columns=['a', 'b', 'c', 'd', 'e'])
frame.cov()
```

```python
frame = DataFrame(randn(20, 3), columns=['a', 'b', 'c'])
frame.ix[:5, 'a'] = np.nan
frame.ix[5:10, 'b'] = np.nan
frame.cov()
```

```python
# Specific the required minimum number of observations
frame.cov(min_periods=12)
```

## Correlation

```python
frame = DataFrame(randn(1000, 5), columns=['a', 'b', 'c', 'd', 'e'])
frame.ix[::2] = np.nan
```

```python
# Normal
frame['a'].corr(frame['b'])
```

```python
# Spearman
frame['a'].corr(frame['b'], method='spearman')
```

```python
# Pairwise correlation of DataFrame columns
frame.corr()
```

```python
# Specific the required minimum number of observations
frame = DataFrame(randn(20, 3), columns=['a', 'b', 'c'])
frame.ix[:5, 'a'] = np.nan
frame.ix[5:10, 'b'] = np.nan
frame.corr()
```

```python
frame.corr(min_periods=12)
```

```python
# Correlation between like-labeled Series contained in DataFrame
index = ['a', 'b', 'c', 'd', 'e']
columns = ['one', 'two', 'three', 'four']
df1 = DataFrame(randn(5, 4), index=index, columns=columns)
df2 = DataFrame(randn(4, 4), index=index[:4], columns=columns)
df1.corrwith(df2)
```

```python
df1.corrwith(df2, axis=1)
```

## Moving and Rolling

```python
ts = Series(randn(1000), index=date_range('1/1/2000', periods=1000))
ts.plot(style='k--')
```

```python
rolling_mean(ts, 60).plot(style='k')
```

```python
df = DataFrame(randn(1000, 4), index=ts.index,
               columns=['A', 'B', 'C', 'D'])
df = df.cumsum()
rolling_sum(df, 60).plot(subplots=True)
```

```python
rolling_sum(df, 60).plot()
```

```python
# Specific function for rolling
mad = lambda x: np.fabs(x - x.mean()).mean()
rolling_apply(ts, 60, mad).plot(style='k')
```

## Predefined rolling function

```python
ser = Series(randn(10), index=date_range('1/1/2000', periods=10))
s1 = rolling_window(ser, 5, 'triang')
s2 = rolling_window(ser, 5, 'boxcar')
s3 = rolling_window(ser, 5, 'gaussian', std=0.1)
df = DataFrame({"traing": s1, "boxcar": s2,
                "qaussian": s3})
df.plot()
```

```python
# Specific labeled center
s1 = rolling_window(ser, 5, 'triang', center=True)
s2 = rolling_window(ser, 5, 'boxcar', center=True)
s3 = rolling_window(ser, 5, 'gaussian', std=0.1, center=True)
df = DataFrame({"traing": s1, "boxcar": s2,
                "qaussian": s3})
df.plot()
```

## Binary rolling moments

```python
df = DataFrame(randn(1000, 4), index=ts.index,
               columns=['A', 'B', 'C', 'D'])
df2 = df[:20]
rolling_corr(df2, df2['B'], window=5)
```

## Computing rolling pairwise covariances and correlations

```python
covs = rolling_cov(df[['B','C','D']], df[['A','B','C']], 50, pairwise=True)
covs[df.index[-50]]
```

```python
correls = rolling_corr(df, 50)
correls[df.index[-50]]
```

```python
correls.ix[:, 'A', 'C'].plot()
```
