# Plotting http://pandas.pydata.org/pandas-docs/stable/visualization.html

```python
from pandas import *
ts = Series(randn(1000), index=date_range("1/1/2000", periods=1000))
ts = ts.cumsum()
ts.plot()
```

```python
df = DataFrame(randn(1000, 4), index=ts.index, columns=list('ABCD'))
df = df.cumsum()
plt.figure();
df.plot();
```

```python
# plot one column versus another using the x and y keywords
df3 = DataFrame(randn(1000, 2), columns=['B', 'C']).cumsum()
df3['A'] = Series(list(range(len(df))))
df3.plot(x='A', y='B')
```

```python
# Bar plots: non-time series data
plt.figure();
df.ix[5].plot(kind='bar');
plt.axhline(0, color='k')
```

```python
# Multi Bar plots
df2 = DataFrame(rand(10, 4), columns=['a', 'b', 'c', 'd'])
df2.plot(kind='bar');
```

```python
# stacked bar plot
df2.plot(kind='bar', stacked=True);
```

```python
# horizontal bar plots
df2.plot(kind='barh', stacked=True);
```

```python
# Histograms
plt.figure();
df['A'].diff().hist()
```

```python
#  histograms of the columns on multiple subplots
df.diff().hist(color='k', alpha=0.5, bins=50)
```

```python
#  plot grouped histograms
data = Series(randn(1000))
data.hist(by=randint(0, 4, 1000), figsize=(6, 4))
```

```python
# Box plots: visualize the distribution of values within each column
df = DataFrame(rand(10,5))
bp = df.boxplot(return_type='dict')
```

```python
# stratified boxplot using the by keyword
df = DataFrame(rand(10,2), columns=['Col1', 'Col2'] )
df['X'] = Series(['A','A','A','A','A','B','B','B','B','B'])
bp = df.boxplot(by='X')
```

```python
# pass a subset of columns to plot
df = DataFrame(rand(10,3), columns=['Col1', 'Col2', 'Col3'])
df['X'] = Series(['A','A','A','A','A','B','B','B','B','B'])
df['Y'] = Series(['A','B','A','B','A','B','A','B','A','B'])
bp = df.boxplot(column=['Col1','Col2'], by=['X','Y'])
```

```python
np.random.seed(1234)
df_box = DataFrame(np.random.randn(50, 2))
df_box['g'] = np.random.choice(['A', 'B'], size=50)
df_box.loc[df_box['g'] == 'B', 1] += 3
bp = df_box.boxplot(by='g')
```

```python
bp = df_box.groupby('g').boxplot()
```

```python
# Area Plots
df = DataFrame(rand(10, 4), columns=['a', 'b', 'c', 'd'])
df.plot(kind='area');
```

```python
# Unstacked area plots
df.plot(kind='area', stacked=False);
```

```python
# Hexagonal Bin Plot: useful alternative to scatter plots 
# if your data are too dense to plot each point individually
df = DataFrame(randn(1000, 2), columns=['a', 'b'])
df['b'] = df['b'] + np.arange(1000)
df.plot(kind='hexbin', x='a', y='b', gridsize=25)
```

```python
# reduces all the values in a bin to a single number
df = DataFrame(randn(1000, 2), columns=['a', 'b'])
df['b'] = df['b'] + np.arange(1000)
df['z'] = np.random.uniform(0, 3, 1000)
df.plot(kind='hexbin', x='a', y='b', C='z', reduce_C_function=np.max,
        gridsize=25)
```

```python
# Lag Plot
from pandas.tools.plotting import lag_plot
plt.figure()
data = Series(0.1 * rand(1000) + 0.9 * np.sin(np.linspace(-99 * np.pi, 99 * np.pi, num=1000)))
lag_plot(data)
```

```python
# macro lag plot by scatter plot
df1 = DataFrame({"a": data, "b": Series(data[1:].tolist() + [0])})
df1
scatter_matrix(df1, alpha=0.2, figsize=(6, 6), diagonal='kde')
```

```python
# Autocorrelation Plot
from pandas.tools.plotting import autocorrelation_plot
data = Series(0.7 * rand(1000) +
              0.3 * np.sin(np.linspace(-9 * np.pi, 9 * np.pi, num=1000)))
autocorrelation_plot(data)
```

```python
# Autocorrelation Plot: Random Data
from pandas.tools.plotting import autocorrelation_plot
data = Series(0.7 * rand(1000))
autocorrelation_plot(data)
```

```python
# Bootstrap Plot
from pandas.tools.plotting import bootstrap_plot
data = Series(rand(1000))
bootstrap_plot(data, size=50, samples=500, color='grey')
```

```python
# Targeting Different Subplots
fig, axes = plt.subplots(nrows=2, ncols=2)
df['A'].plot(ax=axes[0,0]); axes[0,0].set_title('A')
df['B'].plot(ax=axes[0,1]); axes[0,1].set_title('B')
df['C'].plot(ax=axes[1,0]); axes[1,0].set_title('C')
df['D'].plot(ax=axes[1,1]); axes[1,1].set_title('D')
```

```python
# Error Bars
ix3 = MultiIndex.from_arrays([['a', 'a', 'a', 'a', 'b', 'b', 'b', 'b'], ['foo', 'foo', 'bar', 'bar', 'foo', 'foo', 'bar', 'bar']], names=['letter', 'word'])
df3 = DataFrame({'data1': [3, 2, 4, 3, 2, 4, 3, 2], 'data2': [6, 5, 7, 5, 4, 5, 6, 5]}, index=ix3)
gp3 = df3.groupby(level=('letter', 'word'))
means = gp3.mean()
errors = gp3.std()
fig, ax = plt.subplots()
means.plot(yerr=errors, ax=ax, kind='bar')
```

```python
# Plotting Tables
fig, ax = plt.subplots()
df2 = DataFrame(rand(5, 3), columns=['a', 'b', 'c'])
ax.get_xaxis().set_visible(False)
df2.plot(table=True, ax=ax)
```
