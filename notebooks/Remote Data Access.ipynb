## Yahoo! Finance Options

```python
from pandas.io.data import Options
aapl = Options('aapl', 'yahoo')
data = aapl.get_all_data()
data.iloc[0:5, 0:5]
```

```python
#Show the $100 strike puts at all expiry dates:
data.loc[(100, slice(None), 'put'),:].iloc[0:5, 0:5]
```

```python
#Show the volume traded of $100 strike puts at all expiry dates:
data.loc[(100, slice(None), 'put'),'Vol'].head()
```

```python
# Download Partal Data
expiry = datetime.date(2016, 1, 1)
pdata = aapl.get_call_data(expiry=expiry)
pdata.iloc[0:5, 0:5]
```

## World Band

```python
from pandas.io import wb
wb.search('gdp.*capita.*const').iloc[:,:2]
```

```python
dat = wb.download(indicator='NY.GDP.PCAP.KD',
                  country=['US', 'CA', 'MX'],
                  start=2005, end=2008)
print(dat)
```

```python
dat
```

```python
dat['NY.GDP.PCAP.KD'].groupby(level=0).mean()
```

```python
# Compare GDP to the share of people with cellphone contracts
# arround the world
wb.search('cell.*%').iloc[:, :2]
```

```python
ind = ['NY.GDP.PCAP.KD', 'IT.MOB.COV.ZS']
dat = wb.download(indicator=ind, country='all',
                  start=2011, end=2011).dropna()
print dat.tail()
dat.columns = ['gdp', 'cellphone']
print dat.tail()
```

```python
# stats
import statsmodels.formula.api as smf
mod = smf.ols("cellphone ~ np.log(gdp)", dat).fit()
```
